```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("/content/drive/MyDrive/Collab Datasets/DA Exp 6/UniversalBank.csv")
df = df.drop(columns=['ID', 'ZIP Code'], axis=1)
df

# Id and ZIP code columns are removed as they are of no use
```

```python
df.info()
```

```python
fig = px.scatter(df, x='CCAvg', y='Age', color='Personal Loan')
fig.show()
```

```python
df1 = df.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1['Age'] = scaler.fit_transform(df1[['Age']])
df1['Experience'] = scaler.fit_transform(df1[['Experience']])
df1['Income'] = scaler.fit_transform(df1[['Income']])
df1['Mortgage'] = scaler.fit_transform(df1[['Mortgage']])
df1['Family'] = scaler.fit_transform(df1[['Family']])
df1['CCAvg'] = scaler.fit_transform(df1[['CCAvg']])
df1['Education'] = scaler.fit_transform(df1[['Education']])
df = df1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Personal Loan']), df['Personal Loan'], 
                                                    test_size=0.2, random_state=42)
```

```python
from sklearn import svm
clf_svm = svm.SVC(kernel='poly')
clf_svm.fit(X_train, y_train)
y_pred = clf_svm.predict(X_test)
clf_svm.score(X_test, y_test) * 100
```

```python
clf_svm.support_vectors_
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test, y_pred, labels=[0, 1]))
print("="*60)
print(metrics.classification_report(y_test, y_pred, labels=[0, 1]))
```
